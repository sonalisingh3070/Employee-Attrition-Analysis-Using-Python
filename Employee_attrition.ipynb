```python
##Project Title:
Employee Attrition Analysis Using Python: Identifying Key Factors Influencing Employee Turnover
```

```python
##Business Problem: 
Employee attrition is a major challenge for organizations as it leads to increased recruitment costs, loss of experienced 
employees, reduced productivity, and disruptions to business operations.Understanding why employees leave is essential for improving retention 
strategies.
This project analyzes employee data to identify the factors that contribute to attrition and provides data-driven insights that can help HR teams
make informed decisions to improve employee retention.
```

```python
##Project Objectives
The main objectives of this project are:
- Analyze employee attrition patterns using Python.
- Explore the relationship between attrition and employee characteristics such as age, department, salary, overtime, job satisfaction,
  work-life balance, and years at the company.
- Identify the departments and job roles with the highest attrition.
- Discover key factors influencing employee turnover.
- Generate actionable business insights and recommendations to help improve employee retention.
```

```python
##Dataset Overview
- Dataset: IBM HR Employee Attrition
- Total Records : 1470
- Total Features : 35
- Total Variables : Attrition - Yes : Employees left
                                No  : Employees stayed
```

What is the Overall Employee Attrition?

```python
df[df["Attrition"]== "Yes"]["Department"].value_counts()
```

Which Department has the highest Attrition?

```python
department_attrition=(pd.crosstab(df["Department"],df["Attrition"],normalize="index")*100).round(2)
ax=department_attrition.plot(kind="bar",stacked=True,figsize=(8,7),color=("purple","lavender"),edgecolor=("purple"),linewidth=2)
ax.bar_label(ax.containers[0],label_type='center',fmt='%.2f%%')
ax.bar_label(ax.containers[1],label_type='center',fmt='%.2f%%')
plt.title("Department-wise Attrition")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.xticks(rotation=0)
plt.legend(title="Attrition")
plt.show()                                 
```

```python
# The Sales department has the highest number of employees leaving the organization,followed by Human Resources.
```

```python
(pd.crosstab(df["OverTime"],df["Attrition"], normalize="index") *100).round(2)
```

Does Overtime Increase Attrition?

```python
overtime_attrition=(pd.crosstab(df["OverTime"],df["Attrition"],normalize="index")*100).round(2)
ax=overtime_attrition.plot(kind="bar",stacked=True,figsize=(6,5),color=("purple","lavender"),edgecolor=("purple"),linewidth=2)
for container in ax.containers:
    ax.bar_label(container,label_type='center',fmt='%.2f%%')
plt.title("Overtime vs Attrition (%)")
plt.xlabel("Overtime")
plt.ylabel("Percentage of Employees")
plt.xticks(rotation=0)
plt.legend(title="Attrition")
plt.show() 
```

```python
#The Sales department has the highest number of employees leaving compared to other departments.
#Research & Development retains employees more effectively.
```

Business Recommendations...

```python
#Reduce excessive overtime through better workforce planning.
#Improve career growth opportunities in high-turnover departments.
#Conduct employee satisfaction surveys regularly.
#Review compensation for lower-income roles.
#Strengthen onboarding and mentoring for new employees.
```

Conclusion

```python
Employee attrition analysis revealed that employee turnover is influenced by multiple factors rather than a single reason. The analysis showed that
employees working overtime, those with lower job satisfaction, lower monthly income, and fewer years at the company were more likely to leave the 
organization. Higher attrition was also observed in specific departments and job roles, indicating areas where HR should focus its retention 
efforts.
These findings suggest that improving work-life balance, reducing excessive overtime, providing competitive compensation, and creating clear career
growth opportunities can help reduce employee turnover. Regular employee engagement surveys and targeted retention strategies for high-risk groups 
can further improve workforce stability.
Overall, this project demonstrates how data analytics can help organizations identify the key drivers of attrition and support evidence-based HR 
decision-making to improve employee satisfaction and long-term retention.
```
